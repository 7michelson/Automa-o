```python
import numpy as np 
import matplotlib.pyplot as plt
import math
from scipy.constants import h,c, Boltzmann
```

Reproduzindo as constantes físicas importando da biblioteca scipy - apropriadas para cientitas de uma gama variedade de áreas. O lambda varia para vários valores de comprimento de onda, onde a função da lei e Planck em função do comprimento.

```python
pi=np.pi
```

```python
t=6400 #Temperatura do Sol em Kelvin
k= Boltzmann # constatnte de boltzmann
```

```python
Lamb= np.linspace(0.1, 10.1, 80, endpoint=True)
ff=[]
for i in range (0,len(Lamb)):   
    FF=(8*pi*h*c*Lamb[i]**(-5))/ (math.exp((h*c)/(Lamb[i]*k*t)-1))
    ff.append(FF)
plt.figure(figsize=(18,8))
plt.plot(Lamb,ff, 'r.-')
plt.xlabel('Lambda')
plt.ylabel('Função plank')
plt.title(f'Temperatura de {t} k')
    
```

```python
Ray = []
for i in range(0, len(Lamb)):
  Ray.append((pi*8*k*t)/Lamb[i]**4)

plt.figure(figsize=(18,8))
plt.plot(Lamb,Ray, 'r.-')
plt.xlabel('Lambda')
plt.ylabel('Lei de Rayghint')
plt.title(f'Temperatura de {t} k')
```
